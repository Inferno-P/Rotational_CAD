```python
from keras.models import load_model
import h5py
import sys
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load test data
with h5py.File('data.hdf5', 'r') as f:
    X_test = f['X_test'][()]
    Y_test = f['Y_test'][()]

# Add the extra channel axis
X_test = X_test[:,np.newaxis,:,:]
Y_test = Y_test[:, np.newaxis, :, :]
```

# Baseline model

```python
# Import the model
baseline_model = load_model("third_run/flat/saved-model-22-0.0208.hdf5")
```

```python
# Calculate predictions
prediction = baseline_model.predict(x=X_test)

# MSE calculated for each image
mse = ((Y_test-prediction)**2).mean(axis=(1,2,3))
print("Overall MSE: ", mse.mean())

# Plot the MSE distribution as a histogram
plt.figure(figsize=(12,6))
plt.hist(mse, bins=30)
plt.show()
```

```python
# Check out the instance with the highest error
idx = np.where(mse>0.13)[0][0]
print("Index: ", idx)
print("MSE: ", mse[idx])
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,6))

axes[0].imshow(X_test[idx,0,:,:], cmap='Greys_r')
axes[0].set_title("X")

axes[1].imshow(Y_test[idx,0,:,:], cmap='Greys_r')
axes[1].set_title("Y")

axes[2].imshow(prediction[idx,0,:,:], cmap='Greys_r')
axes[2].set_title("Prediction")
plt.show()
```

```python
# Check out the instance with the highest error
idx = np.where(mse>0.12)[0][1]
print("Index: ", idx)
print("MSE: ", mse[idx])
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,6))

axes[0].imshow(X_test[idx,0,:,:], cmap='Greys_r')
axes[0].set_title("X")

axes[1].imshow(Y_test[idx,0,:,:], cmap='Greys_r')
axes[1].set_title("Y")

axes[2].imshow(prediction[idx,0,:,:], cmap='Greys_r')
axes[2].set_title("Prediction")
plt.show()
```
